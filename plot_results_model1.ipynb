```python
import matplotlib.pyplot as plt
```

```python
# load log file
log = open('logs/sentence_model1_1.log', 'r')
lines = log.readlines()
log.close()
```

```python
# save logged results
num_epochs = 54
loss, CER, WER = [], [], []
val_loss, val_CER, val_WER = [], [], []

for i in range(num_epochs):
    line = lines[i + 1]
    line = line.split(' ')
    loss.append(float(line[10][:-1]))
    CER.append(float(line[12][:-1]))
    WER.append(float(line[14][:-1]))
    val_loss.append(float(line[16][:-1]))
    val_CER.append(float(line[18][:-1]))
    val_WER.append(float(line[20]))
```

```python
# plot results
epochs = range(1, num_epochs + 1)
plt.plot(epochs, loss, label='loss')
plt.plot(epochs, val_loss, label='val_loss')
plt.legend()
plt.show()

plt.plot(epochs, CER, label='CER', color='red')
plt.plot(epochs, WER, label='WER', color='blue')
plt.plot(epochs, val_CER, label='val_CER', color='orange', linestyle='--')
plt.plot(epochs, val_WER, label='val_WER', color='green', linestyle='--')
plt.legend()
plt.show()
```
